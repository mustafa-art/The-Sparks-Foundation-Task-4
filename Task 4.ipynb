```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns 
sns.set()
```

```python
# there's 135 columns i'll choose the one that got my intention 

new_data = data[['iyear', 'imonth', 'iday', 'country_txt', 'region_txt', 'provstate', 'city', 'latitude', 'longitude', 'gname',
                'motive', 'attacktype1_txt', 'targtype1_txt', 'target1', 'weaptype1_txt', 'nkill', 'success', 'nwound']]
```

```python
new_data.head()
```

```python
# rename columns 
new_data.rename(columns={'iyear':'Year', 'imonth':'Month', 'iday':'Day', 'country_txt':'Country', 'region_txt':'Region',
                   'provstate':'State', 'city':'City', 'latitude':'Latitude', 'longitude':'Longitude', 
                   'gname':'Group',  'motive':'Motive', 'attacktype1_txt':'Attack_Type', 
                   'targtype1_txt':'Target_Type', 'target1':'Target', 'weaptype1_txt':'Weapon_Type', 
                   'nkill':'Killed', 'success':'Success', 'nwound':'Wounded'}, inplace=True)
```

```python
new_data.head()
```

```python
new_data.info()
```

```python
new_data['Cases'] = new_data['Killed'] + new_data['Wounded']
```

```python
new_data.shape
```

```python
new_data.describe()
```

```python
new_data.head()
```

```python
new_data['Region'].value_counts()
```

```python
x_region = new_data['Region'].value_counts().index
y_region = new_data['Region'].value_counts().values

plt.figure(figsize=(15,8))
sns.barplot(x=x_region, y=y_region)
plt.xticks(rotation=90)
plt.title('Most Region with terrorist attacks')
plt.xlabel('Region')
plt.ylabel('Cases Count')
plt.show()
```

- 'MiddleEast & NorthAfrica' and 'SouthAsia' are the 2 highest regions with terrorist attacks 

```python
# most 10 Countries with terrorist attacks
top_country_attacks = new_data['Country'].value_counts().head(10)
top_country_attacks
```

```python
top_country_attacks.plot(kind='bar', figsize=(6,6))
plt.title('Top 10 country with Terrorist Attacks')
plt.xlabel('Country')
plt.ylabel('Attacks Count')
plt.show()
```

```python
# the most 20 years with biggest cases of injuries+kills with the group and city 
new_data.pivot_table(values='Cases', index=['Year', 'City','Group']).sort_values('Cases', ascending=False).head(20)
```

```python
# let's count cases from terrorist attacks by year  
X= new_data.groupby('Year')['Cases'].sum().index
```

```python
Y = new_data.groupby('Year')['Cases'].sum().values
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=X, y=Y)
plt.xticks(rotation=90)
plt.title('Number of cases by year from Terrorist attacks')
plt.xlabel('Year')
plt.ylabel('Cases Count')
plt.show()
```

- The year with the most cases of injuries+kills from terrorist attacks is 2015 with more than 70000 cases
- we see that the curve is up and down for many times from 1980 to 2010 , in 2012 it increased alot with around 39000, cases and it's just the beginning let's go deeper 

```python
high_cases_years  = new_data[new_data['Year'] >= 2012]
```

```python
high_cases_years.head()
```

```python
# let's see what groups were active at this period of time and remove the first group 'Unknown'
high_cases_years['Group'].value_counts()[1:]
```

- The 'Taliiban' and 'Islamic State of Iraq and the Levant (ISIL)' have made most attacks in the years [2012 >> 2017] with high cases of kills and injuries 

```python
# ok let's make the same pivot table but with the new dataset for the highest cases years 2012 >> 2017  

high_cases_years.pivot_table(values='Cases', index=['Year','Region','Country' , 'City','Group', 'Attack_Type', 'Target_Type', 'Target']).sort_values('Cases', ascending=False).head(30)
```

- United States has the highest number of cases in one attack in years 2012 >> 2017, an armed assault by Anti-Government extremists Group
- Most of the regions here are 'Middle east& NorthAfrica' and 'SouthAsia'>> Afghanistan, with Bombing explosion and armed assault 
- At this period of years was the revoloution of the arabic spring which make sence to produces all these cases on these years 


```python
# let's see the most target type in years 2012 >> 2017
target_wise = high_cases_years.groupby('Target_Type')['Cases'].sum().sort_values(ascending=False)
target_wise
```

```python
target_wise.plot(kind='bar', figsize=(10,8))
plt.title('The most target types that left most casualties for years 2012 to 2017')
plt.ylabel('Casualties Count')
plt.show()
```
